```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
file_path = "D:\\Downloads\\Statement_SES_2011-12-Enrlment.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
print(data.head())

# Ensure the 'year' column is correctly added (if not already done)
if 'Year' in data.columns:
    data['year'] = data['Year'].str[:4].astype(int)
    data.drop(columns=['Year'], inplace=True)

# Group by year and sum the 'All Categories - Class I-V - Total' to get the number of students per year
class_size = data.groupby('year')['All Categories - Class I-V - Total'].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='All Categories - Class I-V - Total', data=class_size)
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.title('Distribution of Students Across Years')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Ensure the 'year' column is correctly added (if not already done)
if 'Year' in data.columns:
    data['year'] = data['Year'].str[:4].astype(int)
    data.drop(columns=['Year'], inplace=True)

# Summing the total number of students in each class group for each year
class_groups = [
    'All Categories - Class I-V - Total',
    'All Categories - Class VI-VIII - Total',
    'All Categories - Class IX-X - Total',
    'All Categories - Class XI-XII - Total'
]

# Summing the total number of students in each class group
class_counts = data[class_groups].sum().reset_index()
class_counts.columns = ['Class Group', 'Total Students']

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Class Group', y='Total Students', data=class_counts)
plt.xlabel('Class Group')
plt.ylabel('Number of Students')
plt.title('Distribution of Students Across Class Groups')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
